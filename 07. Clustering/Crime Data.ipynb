# Importing the Libraries

# Finding out the optimal number of clusters

```python
X = data[['Murder','Assault','Rape','Urbanpop']]
```

```python
data_norm = preprocessing.scale(X)
```

```python
data_norm = pd.DataFrame(data_norm) # standardize the data to normal distribution
```

```python
data_norm.head()
```

```python
plt.figure(figsize=(10,8))
wcss = []
for i in range(1,15):
    kmeans = KMeans(n_clusters=i,init='k-means++', random_state=42)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,15),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### The elbow is at k=4 and let's use it to determine the clusters

# 4 - Analysing the data

```python
kmeans = KMeans(n_clusters=4, init='k-means++',random_state=42)
y_kmeans = kmeans.fit_predict(data_norm)
```

```python
y_kmeans
```

```python
y_kmeans1 = y_kmeans+1
cluster = list(y_kmeans1)
```

```python
data['cluster'] = cluster
```

```python
kmeans_mean_cluster = pd.DataFrame(round(data.groupby('cluster').mean(),1))
kmeans_mean_cluster
```
