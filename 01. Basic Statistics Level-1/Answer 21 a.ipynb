```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Cars.csv')
```

```python
sns.distplot(data.MPG, label='Cars-MPG')
plt.xlabel('MPG')
plt.ylabel('Density')
plt.legend();
```

```python
data.MPG.mean()
```

```python
data.MPG.median()
```

As we can see the mean and median is approximately same therefore, we can say MPG of cars does follow normal distribution.
