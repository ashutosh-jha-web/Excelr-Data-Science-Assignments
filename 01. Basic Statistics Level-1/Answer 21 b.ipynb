```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('wc-at.csv')
```

```python
sns.distplot(data.Waist,label='wc-at-Waist')
plt.xlabel('Waist Circumstances')
plt.ylabel('Density')
plt.legend()
```

```python
data.Waist.mean()
```

```python
data.Waist.median()
```

As we can see the mean and median is approximately same therefore, we can say Waist does follow normal distribution.

```python
sns.distplot(data.AT,label='wc-at-AT')
plt.xlabel('Waist Circumstances')
plt.ylabel('Density')
plt.legend()
```

```python
data.AT.mean()
```

```python
data.AT.median()
```

As we can see the mean and median is not same therefore, we can say Adipose Tissue(AT) does not follow normal distribution.
