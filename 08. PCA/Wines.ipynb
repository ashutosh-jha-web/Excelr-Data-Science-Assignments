# Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler
df_norm = StandardScaler().fit_transform(data) # normalising the data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=13)
```

```python
principalComponents = pca.fit_transform(df_norm)
```

```python
import matplotlib.pyplot as plt
PC = range(1,pca.n_components_+1)
plt.bar(PC, pca.explained_variance_ratio_,color='blue')
plt.xlabel('Principal Components')
plt.ylabel('Variance')
plt.xticks(PC)
```

```python
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0],PCA_components[1],alpha=.3,color='blue')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

## As there is not much variance in 3rd components, so lets just take first 2 componenets for our analysis.

# Finding out the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range(1,15):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(PCA_components.iloc[:,:3])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,15),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Let's use k=3 and determine the clusters

# K Clusters
